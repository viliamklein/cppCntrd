```python
import numpy as np
from astropy.io import fits

import matplotlib.pyplot as plt

cppData = np.loadtxt(fname='cppCntrd.dat', delimiter=',', dtype=np.float64)
eliotData = np.loadtxt(fname='eliotCntrd.dat', delimiter=' ', usecols=(1,2), dtype=np.float64)
```

```python
fig, ax = plt.subplots(2,1, figsize=(8,8))
ax[0].plot(cppData[:,0] - eliotData[:,0])
ax[1].plot(cppData[:,0] - eliotData[:,0])

ax[1].set_xlabel('Index')
ax[0].set_ylabel('$XC_{efy} - XC_{cpp}$')
ax[1].set_ylabel('$YC_{efy} - YC_{cpp}$')

fig.savefig("diffOut.png", dpi=100, )
```
